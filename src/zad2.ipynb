```python
from argon import ArgonSimulation
from matplotlib import pyplot as plt
import pandas as pd
```

```python
params_dict = {
    "n": 6,
    "m": 40.0,
    "e": 1.0,
    "R": 0.38,
    "f": 10000.0,
    "L": 2.418,
    "a": 0.38,
    "T_0": 100,
    "tau": 0.001,
    "S_o": 100,
    "S_d": 10000,
    "S_out": 10,
    "S_xyz": 100
}
```

```python

T_s = [x for x in range(79,90)]
for T in T_s:
    print(f"t = {T}K ")
    out_filepath = f'out/zad2/{T}K.csv'
    xyz_filepath = f'out/zad2/{T}K.xyz'
    params_dict['T_0'] = T
    sim = ArgonSimulation(params_dict, xyz_filepath, out_filepath, show_plots = False)
    sim.run()
```

```python

H_s = []
for T in T_s:
    filepath = f'out/zad2/{T}K.csv'
    df = pd.read_csv(filepath)
    H_s.append(df[' H'].mean())
H_s
plt.plot(T_s, H_s)
plt.xlabel(r'$T_0 \; [K]$')
plt.ylabel(r'$\hat{H}$')

plt.savefig('out/zad2/melting_point.png', dpi=350)
plt.show()
```
